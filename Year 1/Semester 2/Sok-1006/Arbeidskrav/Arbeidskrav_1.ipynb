Oppgave 7.2 A

```python
log_p, log_q = sp.symbols("log_p log_q")

eq = sp.Eq(-2+1.7*log_q, log_p)

supply = sp.solve(eq, log_q)


print(supply)

#np.round(supply)
```

Priselastisiteten til tilbudet er 0.588.
Dette vil si at dersom vi øker prisen med 1%, økes tilbudet med 0.588%. 
Tilbudet er uelastisk. som vil si at tilbudet ikke er følsomt for prisendringer

```python
eq2 = sp.Eq(8.5+0.82*log_q, log_p)

demand = sp.solve(eq2, log_q)

print(demand)
```

Priselastisiteten til etterspørselen er 1.219.
Dette vil si at om vi øker prisen med 1% så reduseres etterspørselen med circa 1.22%.
Etterspørselen er elastisk. som vil si at den er følsom for pris endringer.
