```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-2, 2, 100)
#|x| \leq 2



def funksjon(x):
    return 1.5*x -3


fig, ax = plt.subplots()
#3x - 2y \leq 6
ax.plot(x, funksjon(x), color='black')
ax.fill_between(x, funksjon(x), color='grey', alpha=0.5)
ax.plot(x, 0*x, color='black')
ax.vlines(-2, funksjon(-2), 0, color='black')





#, \>\>\>\> y \leq x, \>\>\>\> 
#plotting in the next line

ax.set_xlim(-4, 4)
ax.set_ylim(-8, 8)

#dashed line at y=0
ax.hlines(0, -4, 4, color='black', linestyle='dashed')
ax.vlines(0, -8, 8, color='black', linestyle='dashed')


plt.show();
```
