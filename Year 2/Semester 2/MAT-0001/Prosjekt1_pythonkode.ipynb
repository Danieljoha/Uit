```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-2, 2, 100)
#|x| \leq 2



def funksjon(x):
    return 1.5*x -3


fig, ax = plt.subplots()
#3x - 2y \leq 6
ax.plot(x, funksjon(x), color='black')
ax.fill_between(x, funksjon(x), color='grey', alpha=0.5)
ax.plot(x, 0*x, color='black')
ax.vlines(-2, funksjon(-2), 0, color='black')





#, \>\>\>\> y \leq x, \>\>\>\> 
#plotting in the next line

ax.set_xlim(-4, 4)
ax.set_ylim(-8, 8)

#dashed line at y=0
ax.hlines(0, -4, 4, color='black', linestyle='dashed')
ax.vlines(0, -8, 8, color='black', linestyle='dashed')


plt.show();
```

```python
punkter = [(-4, 36), (-3, 9), (-2, 2), (-1, 0), (0, 0), (1, 1), (10, 22)]


x = np.linspace(-4.99, 15, 100)
y = 3*x - 12

fig, ax = plt.subplots()

ax.plot(x, 3*x - 12, label='y=3x-12')


for i in punkter:
  ax.scatter(i[0], i[1], color='black')


#\frac{3x(x+1)}{x+5}
def oppgave4(x):
    return 3*x*(x+1)/(x+5)

ax.plot(x, oppgave4(x), label='y=3x(x+1)/(x+5)')

ax.set_xlim(-5, 15)

ax.set_ylim(-5, 50)

ax.hlines(0, -5, 15, color='black', linestyle='dashed')
ax.vlines(0, -5, 50, color='black', linestyle='dashed')

plt.show();
```
